```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

url = "https://drive.google.com/uc?export=download&id=1j35xwKZP2d1QFzVgNu19MXvNDIp0zJBJ"
df = pd.read_csv(url)

print(df.head())
```

```python
# Basic Info 
print(df.info())
print(df.describe(include='all'))
```

```python
# 2. Missing Values 
print("Missing values per column:\n", df.isnull().sum())
```

```python
# 3. Age Distribution 
plt.figure(figsize=(6,4))
sns.countplot(x='Age', data=df)
plt.title("Age Distribution")
plt.show()
```

```python
# 4. Gender Count
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
# 5. College Tier & City Type
plt.figure(figsize=(6,4))
sns.countplot(x='College_Tier', data=df, order=df['College_Tier'].value_counts().index)
plt.title("College Tier Distribution")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(x='City_Type', data=df, order=df['City_Type'].value_counts().index)
plt.title("City Type Distribution")
plt.show()
```

```python
# 6. Monthly Allowance 
print("Average Monthly Allowance:", df['Monthly_Allowance'].mean())
plt.figure(figsize=(6,4))
sns.histplot(df['Monthly_Allowance'], kde=True)
plt.title("Monthly Allowance Distribution")
plt.show()
```

```python
#7. Finance App Usage 
plt.figure(figsize=(6,4))
sns.countplot(y='Uses_Finance_App', data=df, order=df['Uses_Finance_App'].value_counts().index)
plt.title("Finance App Usage")
plt.show()
```

```python
# 8. Part-Time Job Percentage 
part_time_percent = (df['Has_Part_Time_Job'].value_counts(normalize=True) * 100).round(2)
print("Part-Time Job Percentage:\n", part_time_percent)
```

```python
# 9. Top Money Challenges 
top_challenges = df['Top_3_Money_Challenges'].value_counts().head(5)
print("Top 5 Money Challenges:\n", top_challenges)
```

```python
# 10. Cross Analysis 
plt.figure(figsize=(6,4))
sns.barplot(x='Gender', y='Monthly_Allowance', data=df)
plt.title("Monthly Allowance by Gender")
plt.show()

plt.figure(figsize=(6,4))
sns.barplot(x='College_Tier', y='Monthly_Allowance', data=df)
plt.title("Monthly Allowance by College Tier")
plt.show()
```

```python
for col in df.columns:
    print(f"--- {col} ---")
    print(df[col].value_counts())
    print()
```

```python
# 1. Average Monthly Allowance by College Tier
tier_allowance = df.groupby("College_Tier")["Monthly_Allowance"].mean()
print("Average Allowance by College Tier:\n", tier_allowance, "\n")

# 2. Average Monthly Allowance by Gender
gender_allowance = df.groupby("Gender")["Monthly_Allowance"].mean()
print("Average Allowance by Gender:\n", gender_allowance, "\n")

# 3. Part-Time Job counts by Gender
parttime_gender = pd.crosstab(df["Gender"], df["Has_Part_Time_Job"])
print("Part-time Job Counts by Gender:\n", parttime_gender, "\n")

# 4. Finance App usage by Gender
financeapp_gender = pd.crosstab(df["Gender"], df["Uses_Finance_App"])
print("Finance App Usage by Gender:\n", financeapp_gender, "\n")

# 5. Average Monthly Allowance by Age
age_allowance = df.groupby("Age")["Monthly_Allowance"].mean()
print("Average Allowance by Age:\n", age_allowance, "\n")

#  Extra: Loop through all column pairs for numerical comparisons ---
categorical_cols = ["College_Tier", "Gender", "Has_Part_Time_Job", "Uses_Finance_App", "Age"]
numeric_cols = ["Monthly_Allowance"]

for cat in categorical_cols:
    for num in numeric_cols:
        print(f"\nAverage {num} by {cat}:")
        print(df.groupby(cat)[num].mean())
```
